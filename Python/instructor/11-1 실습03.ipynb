(1)  [3, 6, 9, 20, -7, 5] 리스트를 sort 와같은 함수를 사용하지말고 for문을 활용하여 오름차순으로 정렬해주세요.

```python
x = [3, 6, 9, 20, -7, 5]

n = len(x)

for i in range(n):
    for j in range(n-1-i):
        if x[j] > x[j+1]:
            x[j], x[j+1] = x[j+1], x[j]

print(x)
```

버블 정렬 알고리즘이란?
- 리스트의 모든 요소를 순차적으로 비교하고 교환하여 정렬하는 간단한 정렬 알고리즘.
- 매번 가장 큰 요소를 리스트의 끝으로 '버블'시키는 방식으로 작동.

```python
x = [5, 1, 9, 7, 2, 3]

n = len(x) # 리스트의 길이
for i in range(n): # 리스트의 모든 요소가 올바른 위치에 올 때까지 여러 번 순회
    for j in range(0, n-1-i): # 인접한 요소들을 비교하고 교환 (이미 정렬된 마지막 'i'개의 요소를 제외하고 비교를 수행-> 시공간 효율 확보)
        if x[j] > x[j+1]: # 앞의 요소가 뒤의 요소보다 크면
            x[j], x[j+1] = x[j+1], x[j] # 뒤 요소를 교환

# 결과 출력
print(x)
```

```python
x = [5, 1, 9, 7, 2, 3]

n = len(x)  # 리스트의 길이
print(f"초기 x 리스트: {x}")  # 초기 리스트 출력

print(f"---각 단계별 i, j에 따른 x 리스트 변화-----")

for i in range(n):  
    for j in range(0, n-1-i):  # n -1 에 순회한 자리수만큼 빼고 이동
        if x[j] > x[j+1]:  
            x[j], x[j+1] = x[j+1], x[j]  
        print(f"{i}, {j}: {x}")  # 각 비교 후의 리스트 출력

print(f"마지막 x 리스트: {x}")
```

(2) 1-100 까지 숫자중 3과 5의 공배수일경우 “3과 5의 공배수” <br>
나머지 숫자중 3의배수일경우 “3의배수” 나머지 숫자중 5의배수일경우 “5의배수” <br>
모두 해당되지 않을경우 그냥숫자 를 출력하세요

```python
for i in range(1, 101):
    if (i % 3) == 0 & (i % 5) == 0:
        print("3과 5의 공배수")
    elif (i % 3) == 0:
        print("3의 배수")
    elif (i % 5) == 0:
        print("5의 배수")
    else : 
        print(i)
```
