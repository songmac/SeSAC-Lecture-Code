# 데이터 전처리 단계

1. **결측값 처리**
   - 데이터셋에서 결측값을 확인하고 처리하는 단계입니다. 결측값을 삭제하거나 평균, 중앙값, 최빈값 등으로 대체할 수 있습니다. 결측값 처리는 데이터의 완전성을 보장하기 위해 필수적입니다.

2. **데이터 타입 변환**
   - 각 열의 데이터 타입을 확인하고, 필요에 따라 적절한 데이터 타입으로 변환하는 단계입니다. 예를 들어, 문자열로 저장된 숫자를 정수형으로 변환하거나, 날짜 형식의 데이터를 datetime 타입으로 변환합니다.

3. **이상치 처리**
   - 데이터에서 이상치를 식별하고 처리하는 단계입니다. 이상치는 데이터를 왜곡시킬 수 있기 때문에 이를 제거하거나 대체하는 방법을 사용합니다. 주로 IQR(Interquartile Range) 방법을 사용하여 이상치를 제거합니다.
      - IQR: Q3 - Q1, 데이터의 중간 50% 범위
      - 이상치는 일반적으로 Q1 - 1.5 * IQR보다 작거나 Q3 + 1.5 * IQR보다 큰 값으로 정의

4. **데이터 스케일링**
   - 데이터의 범위를 조정하는 단계입니다. 표준화(Standardization)와 정규화(Normalization)를 통해 데이터를 스케일링합니다. 이를 통해 모델이 데이터의 크기 차이에 영향을 받지 않도록 합니다.

5. **범주형 데이터 처리**
   - 범주형 데이터를 숫자형으로 변환하는 단계입니다. 라벨 인코딩(Label Encoding)과 원핫 인코딩(One-Hot Encoding)을 사용하여 범주형 데이터를 처리합니다. 이를 통해 머신러닝 모델이 범주형 데이터를 이해할 수 있도록 합니다.
      - 라벨 인코딩(Label Encoding): 라벨 인코딩은 범주형 데이터를 정수형 숫자로 변환하는 방법. 숫자의 크기에 의미가 있을 때만 적합
      - 원핫 인코딩(One-Hot Encoding): 원핫 인코딩은 범주형 데이터를 이진 벡터로 변환하는 방법. 각 범주를 0과 1로만 구성된 벡터로 변환

6. **특성 생성 및 선택**
   - 새로운 특성을 생성하거나, 중요한 특성을 선택하는 단계입니다. 기존 데이터를 조합하여 새로운 특성을 만들거나, 통계적 방법을 사용하여 중요한 특성을 선택합니다. 이를 통해 모델의 성능을 향상시킬 수 있습니다.

7. **데이터 분할**
   - 데이터를 학습용(train)과 테스트용(test)으로 분할하는 단계입니다. 보통 80:20 또는 70:30 비율로 데이터를 분할합니다. 이를 통해 모델의 성능을 평가하고, 과적합을 방지할 수 있습니다.

8. **데이터 저장**
   - 전처리가 끝난 데이터를 저장하는 단계입니다. CSV, Excel, SQL 등의 형식으로 데이터를 저장하여 이후 분석이나 모델링에 사용할 수 있도록 합니다.

## 연속형 변수의 문제점 및 해결 방법

##### 밴딩이란?
연속형 변수는 머신러닝 모델에서 특정한 문제를 일으킬 수 있습니다. 예를 들어, 연속형 변수는 고유 값의 수가 매우 많아질 수 있어 모델의 복잡성을 증가시키고 과적합을 유발할 수 있습니다. 이러한 문제를 해결하기 위해 연속형 변수를 범주형 변수로 변환하는 방법이 있습니다. 이 과정은 밴딩(Binning) 또는 **정규화(Normalizing)**라고 합니다. </br>
참고로 "Bin"은 사물이나 데이터를 그룹으로 나눈다는 의미입니다.  </br>


#### 밴딩의 종류
- 고정 구간법: 데이터를 일정한 구간으로 나누는 방법
- 등급 구간법: 데이터를 동일한 개수의 구간으로 나누는 방법
- 도메인 기반 구간법: 도메인 지식을 기반으로 데이터를 의미 있는 구간으로 나누는 방법
- 클러스터링 기반 구간법: K-Means와 같은 클러스터링 알고리즘을 사용하여 데이터를 유사한 값들끼리 그룹화하는 방법
- 사용자 정의 함수 (Custom Function): 복잡한 기준이나 다중 조건에 따라 데이터를 범주로 나누고자 할 때 사용

## 원핫 인코딩의 목적 <br>

범주형 변수를 수치형 변수로 변환: 머신러닝 모델은 수치형 데이터를 필요로 하므로 범주형 데이터를 수치형 데이터로 변환해야 합니다. <br>
다중 공선성 방지: 원핫 인코딩은 하나의 범주형 변수를 여러 개의 이진 변수로 변환하여 다중 공선성을 방지합니다. <br>
- 다중 공선성(multicollinearity)
다중 공선성은 회귀 분석에서 설명 변수들 간에 높은 상관관계가 존재하는 경우를 의미합니다. 이는 모델의 예측 성능에 부정적인 영향을 미칠 수 있으며, 회귀 계수의 추정이 불안정해질 수 있습니다.
    - 다중 공선성의 문제점
        - **회귀 계수의 불안정성**: 설명 변수 간의 상관관계가 높으면 회귀 계수가 크게 변동할 수 있습니다.
        - **모델의 해석 어려움**: 어떤 변수가 실제로 중요한지 판단하기 어려워집니다.
        - **모델의 예측 성능 저하**: 다중 공선성은 모델의 일반화 능력을 떨어뜨릴 수 있습니다.

    - 다중 공선성 확인 방법
        - **상관 행렬(correlation matrix)**: 설명 변수들 간의 상관계수를 계산하여 높은 상관관계를 가진 변수들을 식별합니다.
        - **분산 팽창 인자(Variance Inflation Factor, VIF)**: 각 변수의 VIF 값을 계산하여 다중 공선성의 정도를 파악합니다. 일반적으로 VIF 값이 10을 초과하면 다중 공선성이 높다고 판단합니다.

    - 다중 공선성 해결 방법
        - **변수 제거**: 높은 상관관계를 가지는 변수 중 하나를 제거합니다.
        - **주성분 분석(PCA)**: 변수들을 주성분으로 변환하여 상관관계를 줄입니다.
        - **정규화 기법 사용**: 릿지 회귀(Ridge Regression)나 라소 회귀(Lasso Regression)와 같은 정규화 기법을 사용하여 다중 공선성의 영향을 줄일 수 있습니다.

## 승객들 이름을 통해 feature 만들기

Mr(남성)의 경우 여성이나 영유아(Mrs, Miss, Master)의 경우와 극단적인 차이를 보이고있으나, </br> Pclass = 3에서는 남녀 할것 없이 사망율이 높았다는 것은 명백해보인다. </br>

```python
# 새로운 열 'Initial' 생성 및 초기화 (반복문 사용하기 때문에 꼭 명시)
data["Initial"] = 0

# 각 이름에서 이니셜(즉, 'Mr.', 'Mrs.', 'Miss.' 등)을 추출하여 'Initial' 열에 저장
for i in data:
    data["Initial"] = data.Name.str.extract('([A-Za-z]+)\.')

# 추출된 고유 이니셜 값들을 확인
print(data["Initial"].unique())
```

정규 표현식의 의미는 다음과 같다. </br>
'[A-Za-z]' : 대소문자 구분 없이 모든 알파벳 문자를 나타냄 </br>
'+' : 바로 앞 문자 집합 1번 이상 반복 </br>
'\.' : \는 이스케이프 문자(특정 문자를 특별한 의미 없이 그 자체로 인식하게 만들기 위해 사용) 이므로 뒤에 '.'이 오는 패턴을 찾음 </br>
'()' : 추출 또는 캡처되는 부분을 그룹화 </br>

다음은 이니셜들의 뜻과 의미를 정리한 것이다. 각 이니셜들은 주로 이름 앞에 붙여서 사용되며, 사람의 사회적 지위, 직업, 결혼 여부 등을 나타내는 데 사용된다. </br>
Mr (Mister) : 성인 남성에 대한 존칭 </br>
Mrs (Missus): 기혼 여성에 대한 존칭 </br>
Miss (Miss): 미혼 여성에 대한 존칭 </br>
Master (Master): 젊은 남자아이 또는 소년에 대한 존칭 </br>
Don (Don): 스페인어권에서 주로 사용되는 남성 존칭, 특히 존경받는 사람이나 귀족에 사용 </br>
Rev (Reverend): 성직자에 대한 존칭 </br>
Dr (Doctor): 박사 학위 소지자나 의사에 대한 존칭 </br>
Mme (Madame): 프랑스어로 기혼 여성에 대한 존칭 </br>
Ms (Ms): 여성에 대한 일반 존칭, 결혼 여부에 상관없이 사용  </br>
Major (Major): 군대 계급 중 하나로, 소령에 해당  </br>
Lady (Lady): 귀족 여성이나 존경받는 여성에 대한 존칭  </br>
Sir (Sir): 기사 작위를 받은 남성에 대한 존칭, 일반적으로 남성에 대한 공손한 호칭으로도 사용 </br>
Mlle (Mademoiselle): 프랑스어로 미혼 여성에 대한 존칭 </br>
Col (Colonel): 군대 계급 중 하나로, 대령에 해당 </br>
Capt (Captain): 군대나 선박의 대장에 대한 존칭 </br>
Countess (Countess): 백작 부인이나 여성 백작에 대한 존칭 </br>
Jonkheer (Jonkheer): 네덜란드 귀족 계급 중 하나로, 기사에 해당 </br>

```python
# 교차표 생성: Initial과 Sex에 따른 빈도수를 계산
# 결과를 전치(T)하여 성별을 행, 이니셜을 열로 표시
crosstab = pd.crosstab(data.Initial, data.Sex).T
# crosstab = pd.crosstab(data.Sex, data.Initial)

# 스타일 설정: 배경 색상을 'summer_r' 컬러 맵으로 설정하여 그라데이션 적용
styled_crosstab = crosstab.style.background_gradient(cmap='summer_r')

# 스타일링된 교차표 표시
styled_crosstab
```

신기하게도 Initial에 따라서 남녀 구별이 확실하게 된다. (Dr. 의 단 한명의 여자분 빼고!) </br>

영문이름의 이니셜의 의미는 잘 모르지만 성별 구분에 굉장히 강하게 작용할 수 있고 성별 + a의 의미를 가지는 Feature을 만들 수 있을 것으로 보인다. (연령의 의미도 포함한다고 함) </br>

```python
data['Initial'].replace(['Mlle','Mme','Ms','Dr','Major','Lady','Countess','Jonkheer','Col','Rev','Capt','Sir','Don'],
                        ['Miss','Mrs','Miss','Mr','Mr','Mrs','Mrs','Mr','Mr','Other','Mr','Mr','Mr'],inplace=True)
data.groupby(['Initial'])['Age'].mean()
```

```python
data.loc[(data.Age.isnull()) & (data.Initial == 'Master'), 'Age'] = 5
data.loc[(data.Age.isnull()) & (data.Initial == 'Miss'), 'Age'] = 22
data.loc[(data.Age.isnull()) & (data.Initial == 'Mr'), 'Age'] = 33
data.loc[(data.Age.isnull()) & (data.Initial == 'Mrs'), 'Age'] = 36
data.loc[(data.Age.isnull()) & (data.Initial == 'Other'), 'Age'] = 43

data.Age.isnull().any()
```

## 연속형 변수 Age -> Age_binned 로 구간 나누기

```python
# 나이 컬럼의 결측값을 평균값으로 대체 (선택사항)
data['Age'].fillna(data['Age'].mean(), inplace=True)

# 나이 컬럼을 다섯 개의 구간으로 나누기
bins = [0, 12, 20, 40, 60, 80]
labels = ['Child', 'Teenager', 'Adult', 'Middle Aged', 'Senior']
data['Age_binned'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

# 결과 출력
data[['Age', 'Age_binned']].head(10)
# data.head(10)
```

```python
data['Age_binned'].value_counts().to_frame()

# 각 나이 범주별 승객 수를 계산
# .to_frame() --> 시리즈를 데이터프레임으로 변환
```

## SIbSp과 Parch로 Family Size라는 컬럼 만들기

```python
# Family_Size 컬럼 생성 및 계산 (Parch + SibSp)
data['Family_Size'] = data['Parch'] + data['SibSp']  # 부모/자식 수 + 형제/자매 수

# Alone 컬럼 생성 및 계산 (Family_Size가 0이면 Alone은 1)
data['Alone'] = 0
data.loc[data.Family_Size == 0, 'Alone'] = 1 #Family_Size가 0인 행의 Alone 열을 선택해서 값을 1로 지정

data[['Family_Size', 'Alone']].iloc[ :10]
```

## Family_Size와 Survived 관계

```python
# 첫 번째 플롯 생성 (Family_Size와 Survived의 관계)
plt.figure(figsize=(9, 6))
sns.pointplot(
    x='Family_Size',    # x축에 표시할 데이터 열 이름 
    y='Survived',       # y축에 표시할 데이터 열 이름
    data=data,          # 사용할 데이터프레임 
    ci=None             # 에러 바를 표시하지 않음
)
plt.title('Family_Size vs Survived') # 플롯의 제목 설정
plt.show()
```

- 가족 크기(Family_Size)가 1인 경우: 생존율이 약 30% 이상입니다.
- 가족 크기(Family_Size)가 2에서 4인 경우: 생존율이 높아져 약 60%에서 70%에 이릅니다.
- 가족 크기(Family_Size)가 5 이상인 경우: 생존율이 급격히 감소하여, 6명 이상인 경우에는 거의 생존자가 없습니다.

## Family_Size, Pclass와 Survived 관계

## Fare -> Fare Range 로 구간 나누기 

```python
# 요금(Fare) 컬럼의 데이터를 동일한 수(4구간)로 나누어 새로운 컬럼(Fare_Range) 생성
"""
pd.qcut() : qcut은 "quantile cut"의 약자. 이는 데이터의 값을 분위수(quantiles)로 나누는 것을 의미
- 1분위수(Q1): 데이터의 하위 25%
- 2분위수(Q2): 데이터의 하위 50% (중간값)
- 3분위수(Q3): 데이터의 하위 75%
- 4분위수(Q4): 데이터의 상위 25%
"""

data['Fare_Range'] = pd.qcut(data['Fare'], 4)
data['Fare_Range'].head(10)
```

```python
# Fare_Range 별로 생존률의 평균을 계산
fare_survival = data.groupby(['Fare_Range'])['Survived'].mean().to_frame()
fare_survival.style.background_gradient(cmap='summer_r')
```

데이터를 보니 운임료가 올라갈수록 생존율도 높아졌다.

```python
# 요금을 구간으로 나누기 위한 경계값과 구간 레이블 정의
fare_bins = [0, 7.91, 14.454, 31, 513]  # 요금 구간의 경계값
fare_labels = ['Low', 'Medium', 'High', 'Very High']  # 각 구간에 대한 레이블

# 요금(Fare) 데이터를 정해진 구간으로 나누고, 각 구간에 레이블 할당
data['Fare_cat'] = pd.cut(data['Fare'], bins=fare_bins, labels=fare_labels, right=False)

# 결과 출력: 요금과 요금 구간(Fare_cat) 열을 포함하여 상위 10개 행을 출력
data[['Fare_Range', 'Fare_cat']].head(10)
```

```python
sns.catplot(x='Fare_cat', y='Survived', data=data, hue='Sex', kind='point')
plt.show()
```

확실히 Fare을 많이 지불할수록 생존율이 높은 것을 확인할 수 있다.

## Correlation Plot

```python
# 상관 행렬 히트맵 생성
sns.heatmap(data_processed.corr(), annot=True, cmap='RdYlGn', linewidths=0.2, annot_kws={'size': 10})

# 현재 플롯의 Figure 객체 가져오기
fig = plt.gcf()

# Figure 크기 설정
fig.set_size_inches(20, 18)

# x축 및 y축 레이블의 글꼴 크기 설정
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# 그래프 저장
plt.savefig('./result_image/2nd_correlation_heatmap.png', dpi=300, bbox_inches='tight')

# 그래프 표시
plt.show()
```

상관 계수는 -1에서 1 사이의 값을 가지며, 다음을 의미합니다. <br>

- 1: 두 변수 간에 완벽한 양의 상관 관계가 있습니다.
- 0: 두 변수 간에 상관 관계가 없습니다.
- -1: 두 변수 간에 완벽한 음의 상관 관계가 있습니다.

 이 값의 절대값이 클수록 두 변수 간의 상관 관계가 강하다는 것을 의미합니다. <br>

- 0.0 ~ 0.1: 거의 상관 관계가 없음 (negligible correlation)
- 0.1 ~ 0.3: 약한 상관 관계 (weak correlation)
- 0.3 ~ 0.5: 중간 정도의 상관 관계 (moderate correlation)
- 0.5 ~ 0.7: 뚜렷한 상관 관계 (strong correlation)
- 0.7 ~ 1.0: 매우 강한 상관 관계 (very strong correlation)
 
전체 상관 계수 값의 분포를 보고 절대값이 0.3 이상인 변수들을 생존 여부와의 상관관계가 있다고 판단할 수 있습니다.
