# Survived에 영향을 미치는 요소 EDA 분석 및 시각화 보고서

## 1. 데이터 세트 개요
이 데이터 세트는 타이타닉호의 승객 명단에 관한 정보를 포함하고 있으며, 승객의 생존 여부를 중심으로 다양한 특성을 포함하고 있습니다. 데이터는 총 891명의 승객에 대한 정보와 다음의 열을 포함하고 있습니다:

- **PassengerId**: 승객의 고유 식별 번호
- **Survived**: 생존 여부 (0 = 사망, 1 = 생존)
- **Pclass**: 승객의 티켓 클래스
- **Name**: 승객의 이름
- **Sex**: 승객의 성별
- **Age**: 승객의 나이
- **SibSp**: 함께 탑승한 형제자매나 배우자의 수
- **Parch**: 함께 탑승한 부모님 또는 자녀의 수
- **Ticket**: 티켓 번호
- **Fare**: 티켓의 요금
- **Cabin**: 객실 번호
- **Embarked**: 탑승한 항구

## 2. 데이터 정제 및 결측치 처리
데이터 세트에는 일부 열에서 결측치가 발견되었습니다. 주요 결측치는 다음과 같습니다:
- **Age**: 177개의 결측치
- **Cabin**: 687개의 결측치
- **Embarked**: 2개의 결측치

결측치의 처리는 분석의 정확도를 위해 중요합니다. 특히 `Cabin` 열의 결측치는 승객의 생존에 중요한 변수일 수 있으므로, 주의 깊게 다루어야 합니다.

## 4. 초기 가설 설정
- **가설 1**: 나이가 많은 승객은 나이가 적은 승객에 비해 생존률이 낮을 것이다.
- **가설 2**: 1등석 승객은 2등석이나 3등석 승객보다 생존 확률이 높을 것이다.
- **가설 3**: 객실 위치에 따라 생존율이 달라질 것이다.

## 5. 역사학자의 시각에서의 추가 가설
- **가설 4**: "여성과 어린이 먼저"라는 원칙에 따라 여성과 어린이의 생존율이 높을 것이다.
- **가설 5**: 사회적, 경제적 지위가 높은 1등석 승객은 구조 작업에서 우선적으로 고려되어 높은 생존율을 보일 것이다.
- **가설 6**: 승선 위치에 따라 생존율에 차이가 있을 것이다.

## 6. 결론 및 추천
이 분석을 통해 타이타닉의 생존 데이터에 대한 깊은 이해를 얻을 수 있었습니다. 데이터의 정제와 가설 설정을 통해 추가 분석을 진행할 수 있는 기반을 마련하였습니다. 향후 분석에서는 더욱 세부적인 변수의 영향력을 분석하고, 결측치 처리 방법을 개선하여 데이터의 정확성을 높일 필요가 있습니다.

보고서의 시각 자료들은 분석 과정에서 얻은 중요한 인사이트를 효과적으로 전달하는 데 도움을 주었습니다. 추가적인 데이터 분석이나 가설 검증을 위한 진행이 필요할 경우, 적절한 방법론과 도구를 사용하여 보다 정확한 결과를 도출할 수 있을 것입니다.
